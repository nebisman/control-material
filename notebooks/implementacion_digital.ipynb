# Implementación digital de controladores
En este script veremos como implementar digitalmente controladores. Esta primera celda es solo ilustrativa.

```python
import numpy as np
import control as ctrl
 
# Crear función de transferencia en forma zpk (zeros, poles, gain)
# C = zpk([-4.855], [-29.06], 14.989)
C = ctrl.zpk([-4.855], [-29.06], 14.989)
 
# Convertir a espacio de estados
Cve = ctrl.ss(C)
 
# Discretizar con método de Tustin (bilineal)
h = 0.01
Cd = ctrl.c2d(Cve, h, method='tustin')
 
# Mostrar matrices con precisión completa
np.set_printoptions(precision=15)
 
print("A =")
print(Cd.A.flatten())
print("\nB =")
print(Cd.B.flatten())
print("\nC =")
print(Cd.C.flatten())
print("\nD =")
print(Cd.D.flatten())
```
